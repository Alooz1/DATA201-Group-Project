```r
# Add the new countries
The_Netherlands_nzd$country <- "The Netherlands"
Belgium_nzd$country <- "Belgium"
Greece_nzd$country <- "Greece"
Australia_nzd$country <- "Australia"
Japan_nzd$country <- "Japan"
US_nzd$country <- "US"
UK_nzd$country <- "UK"
Spain_nzd$country <- "Spain"
New_Zealand_nzd$country <- "New_Zealand"

# Combine all datasets
combined_data <- rbind(The_Netherlands_nzd, Belgium_nzd, Greece_nzd, Australia_nzd, Japan_nzd, US_nzd, UK_nzd, Spain_nzd, New_Zealand_nzd)

# Clean the data
combined_data$price_nzd <- as.numeric(gsub("[\\$,]", "", combined_data$price_nzd))
```

```r
# Create the plot
ggplot(combined_data, aes(x=reorder(country, price_nzd, FUN=median), y=price_nzd, fill=country)) +
  geom_boxplot(outlier.shape = NA) +
  geom_hline(yintercept = 0, linetype="dashed", color = "red") +
  scale_y_continuous(labels = scales::dollar_format(), breaks = seq(0, 1000, 200), limits = c(0, 1000)) +  # Adjust these values if needed
  coord_trans(y = "sqrt") +
  labs(
    title="The Airbnb Price Distribution in Nine Countries in NZD",
    x="Country",
    y="Price in NZD",
    subtitle="Boxplots represent the distribution of Airbnb listing prices",
    caption="Source: Airbnb"
  ) +
  theme_minimal(base_size = 11) +
  theme(legend.position="none") +
  scale_fill_brewer(palette = "Set3")
```
